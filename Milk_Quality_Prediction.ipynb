```python
from sklearn.preprocessing import LabelEncoder
ec = LabelEncoder()
ec.fit(df['Grade'])
df['Grade'] = ec.transform(df['Grade'])
ec.fit(df['Taste'])
df['Taste'] = ec.transform(df['Taste'])
```

```python
df.head()
```

```python
x = df.drop(['Colour'], axis = 1)
y = df['Colour']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.1, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
```

```python
lr.fit(xtrain,ytrain)
```

```python
dt.fit(xtrain,ytrain)
```

```python
rf.fit(xtrain, ytrain)
```

```python
ypred_lr = lr.predict(xtest)
ypred_dt = dt.predict(xtest)
ypred_rf = rf.predict(xtest)
```

```python
err_lr = ytest - ypred_lr
err_dt = ytest - ypred_dt
err_rf = ytest - ypred_rf
```

```python
print(f'Linear: {np.mean(err_lr)} \n Decision Tree: {np.mean(err_dt)} \n Random Forest: {np.mean(err_rf)}')
```

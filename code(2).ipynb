```python
import numpy as np
x = np.array([[1,2,3,4,5]])
y = np.array([[9,12,27,40,54]])
xmean = np.mean(x)
ymean = np.mean(y)
B1 = np.sum((x - xmean)*(y - ymean))/(np.sum((x-xmean)**2))
print(B1)
B2 = (B1*xmean) - ymean
print(B2)
ypred = B1*x + B2
print(ypred)
```

```python
import random
import numpy as np
import seaborn as sns 
l = [1,3,4,5,9,13,17,21]
c1 = []
c2 = []
m1 = random.choice(l)
while True:
    m2 = random.choice(l)
    if m2 != m1:
        break
while True:
    for i in l:
        if np.abs(m1 - i) > np.abs(m2 - i):
            c2.append(i)
        else:
            c1.append(i)
    new_m1 = np.mean(c1)
    new_m2 = np.mean(c2)
    if new_m1 == m1 and new_m2 ==m2:
        print(f'centroid 1 is {new_m1},centroid is {new_m2}')
        print(f'cluster1 is {c1},cluster is {c2}')
        break
    else:
        m1, m2  =new_m1, new_m2
        c1 = []
        c2 = []
sns.scatterplot(c1)
sns.scatterplot(c2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
inertia = [] #elbow
for k in range (1,10):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit (df)
    inertia.append(model.inertia_)# sum of square of difference in datapoint
print(inertia)
plt.plot(range(1,10),inertia, marker='o')
plt.xlabel('Number of Clusters(k)')
plt.ylabel('Inertia(within cluster sum of squares)')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(df)
ypred = kmeans.predict([[2,3,4,5,3,4,1,8,9,10,11,12]])
if ypred == 1:
    print("motor is working!!")
else:
    print("motor is not working")
ypred
```

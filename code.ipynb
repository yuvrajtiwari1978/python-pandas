```python
import numpy as np
arr = np.array([1, 2, 3, 5, 6, 4, 5])
l = list(arr)
x = 0
for i in arr:
    c = 0
    for j in arr:
        c= l.count(i)
        if c > x:
            x = c
            e=i
print(e,x)
print(np.max(arr))
print(np.min(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(f'range: {np.max(arr) - np.min(arr)}')
```

```python
import random

class ladders:
    def __init__(self, count):
        self.count = count
        self.sum = 0
        while self.sum != 100:
            self.count += 1
            self.temp = self.sum
            self.temp += random.choice([1,2,3,4,5,6])
            if self.temp == 89 or self.temp == 97:
                self.temp -= 14 
                self.sum = self.temp
            if self.temp == 9 or self.temp == 1:
                self.temp +=91
                self.sum = self.temp
            if self.temp == 100 or self.temp<100:
                self.sum = self.temp
        print(self.count)
obj = ladders
print(obj(0))
```
